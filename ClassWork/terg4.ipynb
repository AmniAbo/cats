```python
import matplotlib.pyplot as plt
import json
from google.colab import drive

# Mount Google Drive to access the JSON file
drive.mount('/content/drive')

# Path to your JSON file
json_file = '/content/drive/My Drive/data.json'

# Load JSON data
with open(json_file, 'r') as f:
    data = json.load(f)

# Extracting data for plotting
xAxis = [entry['Description'] for entry in data]
yAxis = [entry['Time'] for entry in data]

# Plotting
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.grid(True)

# LINE GRAPH
plt.plot(xAxis, yAxis, color='#8a5082', marker='o')
plt.xlabel('Description')
plt.ylabel('Time')
plt.title('Time vs Description')

plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

plt.tight_layout()
plt.show()

```
